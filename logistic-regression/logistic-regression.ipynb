### Understanding the odds and log odds
The sigmoid function provides the probability of the obervation to belong to a particular class given the univariate x. Thus,

p = 1/(1 + e<sup>-(B<sub>0</sub>+B<sub>1</sub>*X)</sup>) <br>
1-p = e<sup>-(B<sub>0</sub>+B<sub>1</sub>*X)</sup>) / (1 + e<sup>-(B<sub>0</sub>+B<sub>1</sub>*X)</sup>) <br>

thus the odds is defined as:
p/(1-p) = e<sup>(B<sub>0</sub>+B<sub>1</sub>*X)</sup>) <br>

thus 
log(p/(1-p)) = (B<sub>0</sub>+B<sub>1</sub>*X)</sup> <br>
